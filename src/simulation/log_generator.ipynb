```python
import time
import random
import requests
from datetime import datetime, UTC

MASTER_URL = "https://unlikably-unremissible-yamileth.ngrok-free.dev/receive_log"  # ngrok URL du master

ATTACK_CLASSES = ["SQLi", "XSS", "CommandInjection", "PathTraversal", "SuspiciousUA"]
NORMAL_CLASS = "Normal"

SQL_PAYLOADS = [
    "' OR 1=1 --",
    "' UNION SELECT * FROM users --",
    "1' OR '1'='1",
    "'; DROP TABLE users; --",
    "\" OR 1=1 /*"
]

CMD_PAYLOADS = [
    "test; ls -la",
    "image.png && cat /etc/passwd",
    "name=`whoami`",
    "ping 8.8.8.8; rm -rf /"
]

XSS_PAYLOADS = [
    "<script>alert(1)</script>",
    "<img src=x onerror=alert(1)>",
    "<svg/onload=alert('XSS')>"
]

PATH_PAYLOADS = [
    "../etc/passwd",
    "../../../../../windows/system32",
    "/../../../../var/log/auth.log",
    "../admin/secrets"
]

SUSPICIOUS_AGENTS = [
    "sqlmap/1.5",
    "nmap script",
    "curl exploit",
    "attack-bot"
]

BENIGN_AGENTS = [
    "Mozilla/5.0",
    "Chrome/122.0",
    "Safari/15.1",
    "Edge/101.0",
    "Android WebView"
]

ENDPOINTS = [
    "/login", "/auth", "/search", "/api/data", "/download/file",
    "/comment/add", "/profile", "/upload", "/products", "/admin"
]


def random_ip():
    return f"{random.randint(1, 255)}.{random.randint(0, 255)}.{random.randint(0, 255)}.{random.randint(1, 255)}"


def generate_log():
    """Génère un log réaliste"""
    attack_type = "Normal" if random.random() < 0.8 else random.choice(ATTACK_CLASSES)

    endpoint = random.choice(ENDPOINTS)
    ip = random_ip()
    method = random.choice(["GET", "POST"])

    if attack_type == "SQLi":
        payload = random.choice(SQL_PAYLOADS)
        user_agent = random.choice(BENIGN_AGENTS)

    elif attack_type == "CommandInjection":
        payload = random.choice(CMD_PAYLOADS)
        user_agent = random.choice(BENIGN_AGENTS)

    elif attack_type == "XSS":
        payload = random.choice(XSS_PAYLOADS)
        user_agent = random.choice(BENIGN_AGENTS)

    elif attack_type == "PathTraversal":
        payload = random.choice(PATH_PAYLOADS)
        user_agent = random.choice(BENIGN_AGENTS)

    elif attack_type == "SuspiciousUA":
        payload = ""
        user_agent = random.choice(SUSPICIOUS_AGENTS)

    else:  # Normal
        payload = ""
        user_agent = random.choice(BENIGN_AGENTS)

    log = {
        "timestamp": datetime.now(UTC).isoformat().replace('+00:00', 'Z'),
        "method": method,
        "endpoint": endpoint,
        "user_agent": user_agent,
        "payload": payload,
        "ip": ip,
        "real_category": attack_type
    }

    try:
        requests.post(MASTER_URL, json=log)
        print("Log envoyé:", log)
    except Exception as e:
        print("Erreur d'envoi au Master:", e)

    time.sleep(1)  # 1 log par seconde

    return log


# Boucle continue pour générer les logs
if __name__ == "__main__":
    while True:
        generate_log()
```
